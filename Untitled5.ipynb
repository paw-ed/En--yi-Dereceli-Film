```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
data = pd.read_csv("movie.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
data['release_date'] = pd.to_datetime(data['release_date'], errors = 'coerce')
```

```python
data.isnull().sum()
```

```python
for i in data.select_dtypes(include = 'object').columns:
    data['overview'].fillna(data['overview'].mode()[0], inplace = True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.columns
```

```python
data.describe(include = 'number').T
```

```python
sns.pairplot(data)
plt.show()
```

```python
for i in data.select_dtypes(include = 'number').columns:
    sns.histplot(data = data, x = i, kde = True, color = 'blue')
    plt.title(f'{i}')
    plt.show()
```

```python
for i in data.select_dtypes(include = 'number').columns:
    fig = px.box(data, x = i)
    fig.update_layout(
    title = f'{i}',
    height = 300,
    width = 600
    )
    fig.show()
```

```python
fig = px.scatter(data, x = 'popularity',
                y = 'vote_count',
                title = 'Vote Average vs Popularity',
                hover_data = ['title'])
fig.show()
```

```python
fig = px.scatter(data,
                 x = 'vote_count',
                 y = 'vote_average',
                 size = 'popularity',
                 color = 'popularity',
                  title = 'Vote Count vs Vote Average',
                 hover_data = ['title'])
fig.show()
```
